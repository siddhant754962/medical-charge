```python
import matplotlib.pyplot as plt
import seaborn as sns

# Shape of dataset
print("Dataset shape:", data.shape)

# Null values check
print("null values :" ,data.isnull().sum().sum())

# Duplicate rows check
print("Duplicate rows:", data.duplicated().sum())
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data['charges'], bins=15, kde=True)
plt.title("Distribution of Medical Charges")
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
sns.histplot(data['age'], bins=30, kde=True, color="blue")
plt.title("Age Distribution")

plt.subplot(1,3,2)
sns.histplot(data['bmi'], bins=30, kde=True, color="green")
plt.title("BMI Distribution")

plt.subplot(1,3,3)
sns.countplot(x='children', data=data, palette="Set2")
plt.title("Children Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,3,1)
sns.countplot(x='sex', data=data, palette="Set1")
plt.title("Gender Distribution")

plt.subplot(1,3,2)
sns.countplot(x='smoker', data=data, palette="Set2")
plt.title("Smoker Distribution")

plt.subplot(1,3,3)
sns.countplot(x='region', data=data, palette="Set3")
plt.title("Region Distribution")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='age', y='charges', hue='smoker', data=data)
plt.title("Age vs Charges (Smoker vs Non-Smoker)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=data)
plt.title("BMI vs Charges (Smoker vs Non-Smoker)")
plt.show()
```
